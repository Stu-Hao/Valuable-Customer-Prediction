```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv('SALES_DATA.csv')
```

```python
data.info()
data.head()
```

```python
data=data[data['Invoice Date']!='2/29/2018']
```

```python
data['Invoice Date'] = pd.to_datetime(data['Invoice Date'])
```

```python
# check number of orders per customer
freqdata=data[['CustomerID','InvoiceNo']].drop_duplicates().sort_values('CustomerID')

freqdata.info()
freqdata.head(10)
```

```python
ordercount=freqdata.groupby('CustomerID')['InvoiceNo'].count()

ordercount.head(10)
```

```python
# check number of unique invoice date per customer
id_date=data[['CustomerID','Invoice Date']].drop_duplicates().sort_values('CustomerID')

id_date.info()
id_date.head(10)
```

```python
datecount=id_date.groupby('CustomerID')['Invoice Date'].count()

datecount.head(10)
```

```python
# get sales per year for each customer (2016 and 2017 are Year 1)
yearly=data[['CustomerID','Year Idx','Ext-Price']].sort_values('CustomerID')

yearly.info()
yearly.head()
```

```python
yeargroup=yearly.groupby(['CustomerID','Year Idx'])['Ext-Price'].aggregate('sum').reset_index()

yeargroup.info()
yeargroup.head()
```

```python
yeargroup.to_csv('Sales per Year.csv',index=True)
```

```python
# get sales per half year for each customer (the 4th period got 7 monthes)
half_year=data[['CustomerID','Half Year Idx','Ext-Price']].sort_values('CustomerID')

half_year.info()
half_year.head()
```

```python
halfyeargroup=half_year.groupby(['CustomerID','Half Year Idx'])['Ext-Price'].aggregate('sum').reset_index()

halfyeargroup.info()
halfyeargroup.head()
```

```python
halfyeargroup.to_csv('Sales per Half Year.csv',index=True)
```
