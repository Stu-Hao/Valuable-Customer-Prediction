```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Sales Data 2018.csv')
```

```python
data.info()
data.head()
```

```python
np.percentile(data['Invoice Date'],50)
```

```python
newdata=data[['CustomerID','InvoiceNo','Invoice Date','Month']].drop_duplicates().sort_values('CustomerID')

newdata.info()
newdata.head()
```

```python
# check number of orders per customer
freqdata=data[['CustomerID','InvoiceNo']].drop_duplicates().sort_values('CustomerID')

freqdata.info()
freqdata.head(10)
```

```python
ordercount=freqdata.groupby('CustomerID')['InvoiceNo'].count()

ordercount.head(10)
```

```python
# check number of unique invoice date per customer
id_date=data[['CustomerID','Invoice Date']].drop_duplicates().sort_values('CustomerID')

id_date.info()
id_date.head(10)
```

```python
datecount=id_date.groupby('CustomerID')['Invoice Date'].count()

datecount.head(10)
```

```python
# get 5 dates for each customer
id_date['Date1']=id_date['Invoice Date']
id_date['Date2']=id_date['Invoice Date']
id_date['Date3']=id_date['Invoice Date']
id_date['Date4']=id_date['Invoice Date']
id_date['Date5']=id_date['Invoice Date']
```

```python
byorder=id_date.groupby('CustomerID').agg({'Date1': lambda x: x.min(),'Date2': lambda x: round(np.percentile(x,25)),
                                           'Date3': lambda x: round(np.percentile(x,50)),'Date4': lambda x: round(np.percentile(x,75)),
                                           'Date5': lambda x: x.max()})

byorder.info()
byorder.head()
```

```python
byorder.to_csv('Five Dates.csv',index=True)
```

```python
# check if every invoiceNo is connected to a single date
order_date=data[['InvoiceNo','Invoice Date']].drop_duplicates().sort_values('InvoiceNo')

order_date.info()
order_date.head(10)
```

```python
order_date_count=order_date.groupby('InvoiceNo').agg({'Invoice Date': lambda x: x.count()})

order_date_count.info()
order_date_count.head(10)
```

```python
order_date_count.to_csv('order_date_count.csv',index=True)
```

```python
# get panel data of monthly sales per customer
monthly=data[['CustomerID','Month','Ext-Price']].sort_values('CustomerID')

monthly.info()
monthly.head()
```

```python
monthgroup=monthly.groupby(['CustomerID','Month']).agg({'Ext-Price': lambda x: x.sum()})

monthgroup.info()
monthgroup.head()
```

```python
group=monthly.groupby(['CustomerID','Month'])['Ext-Price'].aggregate('sum').reset_index()

group.info()
group.head()             
```

```python
group.to_csv('Monthly Sales.csv',index=True)
```

```python
# export number of orders and dates
ordercount.to_csv('Order Count.csv',index=True)
datecount.to_csv('Date Count.csv',index=True)
```

```python
# get panel data of sales per order per customer
per_order=data[['CustomerID','InvoiceNo','Ext-Price']].sort_values('CustomerID')

per_order.info()
per_order.head(10)
```

```python
ordergroup=per_order.groupby(['CustomerID','InvoiceNo'])['Ext-Price'].aggregate('sum').reset_index()

ordergroup.info()
ordergroup.head(10)
```

```python
ordergroup.to_csv('Sales per Order.csv',index=True)
```

```python
# get panel data of sales per date per customer
per_date=data[['CustomerID','Invoice Date','Ext-Price']].sort_values('CustomerID')

per_date.info()
per_date.head(10)
```

```python
dategroup=per_date.groupby(['CustomerID','Invoice Date'])['Ext-Price'].aggregate('sum').reset_index()

dategroup.info()
dategroup.head(10)
```

```python
dategroup.to_csv('Sales per Date.csv',index=True)
```
